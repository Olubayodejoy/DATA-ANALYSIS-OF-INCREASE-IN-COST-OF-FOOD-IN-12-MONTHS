```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': ['October 2022', 'November 2022', 'December 2022', 'January 2023', 'February 2023', 'March 2023',
              'April 2023', 'May 2023', 'June 2023', 'July 2023', 'August 2023', 'September 2023'],
    'Food Cost Increase (%)': [10.1, 10.6, 10.0, 9.4, 9.7, 8.9, 9.2, 8.7, 7.5, 7.0, 6.8, 6.2]
}

cost2_df = pd.DataFrame(data)
```

```python
cost2_df
```

```python
plt.figure(figsize=(10,6))
plt.plot(cost2_df['Month'], cost2_df['Food Cost Increase (%)'], marker='o', color='b')

plt.title('Monthly Food Cost Increase in Canada (2022-2023)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Food Cost Increase (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
sns.barplot(x='Month', y='Food Cost Increase (%)', data=cost2_df)
plt.show()
```

```python

plt.figure(figsize=(8,8))
plt.pie(cost2_df['Food Cost Increase (%)'], labels=cost2_df['Month'], autopct='%1.1f%%', startangle=140)

plt.title('Food Price Increases in Canada (2023)', fontsize=14)

plt.show()
```
